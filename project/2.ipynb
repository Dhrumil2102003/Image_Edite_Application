### Image Dimensions (in pixels)

* image is considered to be height and width in pixels
* the coordinates system here starts in the upper left corner
* the starting point of the image is x and y axis is the upper left corner


### Crop the Image in python 
1. Load the image
2. Get dimensions of the image
3. Determine what you want to cropped
4. Create a new Image based on the dimensions
5. Write new image

```python
import cv2

def main():
    img = cv2.imread("C:/Users/ADMIN/Music/project/image/img/1.jpeg")
    dimensions = img.shape
    height = dimensions[0]  
    width = dimensions[1]       
    cropped_image = img[200:(height-200), 200:(width-400)]
    cv2.imwrite("Cropped1.jpg", cropped_image)


if __name__ == "__main__": # It is also possible to use this syntax in python code that we want to run only when the script is executed, not when it is imported into another script or program or when it is run as a stand-alone program.
    main()
```

```python
import tkinter as tk
import tkinter.filedialog as tk_file 
from PIL import Image,ImageTk


root = tk.Tk()
root.geometry('800x600')
root.config(bg='black')

open_image = ''
img_path = '' 

edited_image=''

image_height = 0
image_width = 0

start_x = start_y = 0  
rect_id = None

def save_image():
    file_name = tk_file.asksaveasfilename(defaultextension=".png", filetypes=[("PNG files", "*.png"), 
                                                                              ("JPEG files", "*.jpg"), 
                                                                              ("All files", "*.*")])
    
    if file_name:
        edited_image.save(file_name )


def clear_image():
    global open_image, edited_image, img_path, image_lb

    if img_path:
        open_image = ImageTk.PhotoImage(Image.open(img_path))
        edited_image = Image.open(img_path)  
        image_lb.config(image=open_image)


def crop_image_with_mouse(left,top,right,bottom):
    global edited_image, open_image, image_width,image_height
    
    edit_img = Image.open(img_path)
    # in PIL module image.crop() it's take tupel argument in tupel we will pass crop area of image (left,top,image_widrh-right,image_width-bottom)
    edited_image = edit_img.crop((left,top,image_width-right,image_height-bottom))
    open_image = ImageTk.PhotoImage(edited_image)
    image_lb.config(image=open_image)
    
def open():
    global open_image, img_path,image_height,image_width

    img_path = tk_file.askopenfilename()
    if img_path:
        open_image = ImageTk.PhotoImage(Image.open(img_path))
        image_lb.config(image=open_image)
        image_height = open_image.height()
        image_width = open_image.width()


def start_crop(event):
    global start_x, start_y, rect_id
    start_x, start_y = event.x, event.y
    if rect_id:
        canvas.delete(rect_id)  # Remove the previous rectangle if present
    rect_id = canvas.create_rectangle(start_x, start_y, start_x, start_y, outline='red')


def end_crop(event):
    global start_x, start_y, rect_id
    # Get the coordinates of the rectangle
    x1, y1, x2, y2 = start_x, start_y, event.x, event.y
    # Ensure coordinates are within bounds and ordered correctly
    left = min(x1, x2)
    right = max(x1, x2)
    top = min(y1, y2)
    bottom = max(y1, y2)
    
    crop_image_with_mouse(left, top, right, bottom)   

def control_windows(w, h):
    controls = tk.Toplevel(root)
    controls.geometry('300x300')
    
    left_lb = tk.Label(controls, text = 'Left')
    left_lb.pack(anchor = tk.W, pady = 5)
    
    left_scale = tk.Scale(controls, from_=0, to = w, orient=tk.HORIZONTAL,
                          command=lambda x: crop_image_with_mouse(left_scale.get(), 
                                                           top_scale.get(), 
                                                           right_scale.get(), 
                                                           bottom_scale.get()))
    left_scale.pack(anchor=tk.W, fill=tk.X)
    
    right_lb = tk.Label(controls, text = 'Right')
    right_lb.pack(anchor = tk.W, pady = 5)
    
    right_scale = tk.Scale(controls, from_=0, to = w, orient=tk.HORIZONTAL,
                           command=lambda x: crop_image_with_mouse(left_scale.get(), 
                                                           top_scale.get(), 
                                                           right_scale.get(), 
                                                           bottom_scale.get()))
    right_scale.pack(anchor=tk.W, fill=tk.X)
    
    top_lb = tk.Label(controls, text = 'Top')
    top_lb.pack(anchor = tk.W, pady = 5)
    
    top_scale = tk.Scale(controls, from_=0, to = h, orient=tk.HORIZONTAL,
                         command=lambda x: crop_image_with_mouse(left_scale.get(), 
                                                           top_scale.get(), 
                                                           right_scale.get(), 
                                                           bottom_scale.get()))
    top_scale.pack(anchor=tk.W, fill=tk.X)
    
    bottom_lb = tk.Label(controls, text = 'Bottom')
    bottom_lb.pack(anchor = tk.W, pady = 5)
    
    bottom_scale = tk.Scale(controls, from_=0, to = h, orient=tk.HORIZONTAL,
                            command=lambda x: crop_image_with_mouse(left_scale.get(), 
                                                           top_scale.get(), 
                                                           right_scale.get(), 
                                                           bottom_scale.get()))
    bottom_scale.pack(anchor=tk.W, fill=tk.X)
    
    controls.mainloop()

menu_bar = tk.Menu(root)
menu_bar.add_command(label='Open', command=open)
menu_bar.add_command(label='Save',command=save_image)
menu_bar.add_command(label='Clear',command=clear_image)

root.config(menu=menu_bar)

canvas = tk.Canvas(root, bg='gray')
canvas.pack(fill=tk.BOTH, expand=True)

canvas.bind("<ButtonPress-1>", start_crop)
canvas.bind("<ButtonRelease-1>", end_crop)

image_lb = tk.Label(root, bg='gray')
image_lb.pack(fill=tk.BOTH)

root.mainloop()
```

```python
from tkinter import *
import tkinter as tk
import tkinter.filedialog as tk_file
from PIL import Image, ImageTk

root = tk.Tk()
root.geometry('1000x600')
root.config(bg='black')

open_image = ''
img_path = ''
edited_image = ''
image_height = 0
image_width = 0
start_x = start_y = 0  
rect_id = None 

def save_image():
    file_name = tk_file.asksaveasfilename(defaultextension=".png", filetypes=[("PNG files", "*.png"), 
                                                                              ("JPEG files", "*.jpg"), 
                                                                              ("All files", "*.*")])
    
    if file_name:
        edited_image.save(file_name)

def clear_image():
    global open_image, edited_image, img_path

    if img_path:
        open_image = ImageTk.PhotoImage(Image.open(img_path))
        edited_image = Image.open(img_path)  
        canvas.delete("all")  
        canvas.create_image(0, 0, anchor=tk.NW, image=open_image)

def crop_image_with_mouse(left, top, right, bottom):
    global edited_image, open_image
    
    edit_img = Image.open(img_path)
    edited_image = edit_img.crop((left, top, right, bottom))
    open_image = ImageTk.PhotoImage(edited_image)
    canvas.delete("all")  
    canvas.create_image(0, 0, anchor=tk.NW, image=open_image)

def open_image_file():
    global open_image, img_path, image_height, image_width

    img_path = tk_file.askopenfilename()
    if img_path:
        opened_image = Image.open(img_path)
        image_height = opened_image.height
        image_width = opened_image.width
        open_image = ImageTk.PhotoImage(opened_image)
        canvas.config(width=image_width, height=image_height)
        canvas.create_image(0, 0, anchor=tk.NW, image=open_image)

def start_crop(event):
    global start_x, start_y, rect_id
    start_x, start_y = event.x, event.y
    if rect_id:
        canvas.delete(rect_id) 
    rect_id = canvas.create_rectangle(start_x, start_y, start_x, start_y, outline='red')

def update_crop_area(event):
    global rect_id
    canvas.coords(rect_id, start_x, start_y, event.x, event.y)

def end_crop(event):
    global start_x, start_y, rect_id
    x1, y1, x2, y2 = start_x, start_y, event.x, event.y
    left = min(x1, x2)
    right = max(x1, x2)
    top = min(y1, y2)
    bottom = max(y1, y2)
    
    crop_image_with_mouse(left, top, right, bottom)

# Menu bar for actions
menu_bar = tk.Menu(root)
menu_bar.add_command(label='Open', command=open_image_file)
menu_bar.add_command(label='Save', command=save_image)
menu_bar.add_command(label='Clear', command=clear_image)

root.config(menu=menu_bar)

canvas = tk.Canvas(root, bg='gray')
canvas.pack(fill=tk.BOTH, expand=True)

canvas.bind("<ButtonPress-1>", start_crop)
canvas.bind("<B1-Motion>", update_crop_area)
canvas.bind("<ButtonRelease-1>", end_crop)


root.mainloop()
```

```python
from tkinter import *
import tkinter as tk
import tkinter.filedialog as tk_file
from tkinter.colorchooser import askcolor
from PIL import Image, ImageTk, ImageDraw

root = tk.Tk()
root.geometry('1000x600')
root.config(bg='black')

open_image = ''
img_path = ''
edited_image = None
image_height = 0
image_width = 0
start_x = start_y = 0  
rect_id = None 
brush_color = 'black'
brush_size = 5
drawing = False  
current_mode = 'crop'  

def save_image():
    file_name = tk_file.asksaveasfilename(defaultextension=".png", filetypes=[("PNG files", "*.png"), 
                                                                              ("JPEG files", "*.jpg"), 
                                                                              ("All files", "*.*")])
    if file_name:
        edited_image.save(file_name)


def clear_image():
    global open_image, edited_image, img_path

    if img_path: 
        opened_image = Image.open(img_path)
        edited_image = opened_image.copy()  
        open_image = ImageTk.PhotoImage(opened_image) 
        canvas.delete("all")  
        canvas.create_image(0, 0, anchor=tk.NW, image=open_image)  


def open_image_file():
    global open_image, img_path, image_height, image_width, edited_image

    img_path = tk_file.askopenfilename()
    if img_path:
        opened_image = Image.open(img_path)
        edited_image = opened_image.copy()  
        image_height = opened_image.height
        image_width = opened_image.width
        open_image = ImageTk.PhotoImage(opened_image)
        canvas.config(scrollregion=canvas.bbox("all"))  
        canvas.delete("all")
        canvas.create_image(0, 0, anchor=tk.NW, image=open_image)


def select_color():
    global brush_color
    color = askcolor()[1]  
    if color:
        brush_color = color


def set_brush_size(value):
    global brush_size
    brush_size = int(value)


def start_drawing(event):
    global drawing, start_x, start_y
    if current_mode == 'brush':
        drawing = True
        start_x, start_y = event.x, event.y  


def draw_brush(event):
    global drawing, edited_image, start_x, start_y

    if drawing and current_mode == 'brush':
        x, y = event.x, event.y
        
        # Draw on the canvas
        canvas.create_line(start_x, start_y, x, y, fill=brush_color, width=brush_size, capstyle=ROUND, smooth=True)

        # Draw on the image
        if edited_image:
            draw = ImageDraw.Draw(edited_image)
            draw.line([start_x, start_y, x, y], fill=brush_color, width=brush_size)

        start_x, start_y = x, y  


def stop_drawing(event):
    global drawing
    drawing = False  


def start_crop(event):
    global start_x, start_y, rect_id
    if current_mode == 'crop':
        start_x, start_y = event.x, event.y
        if rect_id:
            canvas.delete(rect_id) 
        rect_id = canvas.create_rectangle(start_x, start_y, start_x, start_y, outline='red')


def update_crop_area(event):
    global rect_id
    if current_mode == 'crop':
        canvas.coords(rect_id, start_x, start_y, event.x, event.y)


def end_crop(event):
    global start_x, start_y, rect_id
    if current_mode == 'crop':
        x1, y1, x2, y2 = start_x, start_y, event.x, event.y
        left = min(x1, x2)
        right = max(x1, x2)
        top = min(y1, y2)
        bottom = max(y1, y2)
        
        crop_image_with_mouse(left, top, right, bottom)


def crop_image_with_mouse(left, top, right, bottom):
    global edited_image, open_image
    
    edited_image = edited_image.crop((left, top, right, bottom))
    open_image = ImageTk.PhotoImage(edited_image)
    canvas.delete("all")  
    canvas.create_image(0, 0, anchor=tk.NW, image=open_image)


def toggle_mode(mode):
    global current_mode
    current_mode = mode
    if mode == 'brush':
        frame1.pack(side=tk.TOP, fill=tk.X)  
        canvas.config(cursor="pencil")  
    elif mode == 'crop':
        frame1.pack_forget()  
        canvas.config(cursor="crosshair")  


# Menu bar for actions
menu_bar = tk.Menu(root)
menu_bar.add_command(label='Open', command=open_image_file)
menu_bar.add_command(label='Save', command=save_image)
menu_bar.add_command(label='Clear', command=clear_image)
menu_bar.add_command(label='Brush', command=lambda: toggle_mode('brush'))  
menu_bar.add_command(label='Crop', command=lambda: toggle_mode('crop'))  

root.config(menu=menu_bar)


frame1 = Frame(root, height=100, bg="lightgray")

brush_size_label = Label(frame1, text="Brush Size:", bg="lightgray")
brush_size_label.pack(side=tk.LEFT, padx=10)

brush_size_slider = Scale(frame1, from_=1, to=20, orient=HORIZONTAL, command=set_brush_size)
brush_size_slider.pack(side=tk.LEFT, padx=10)

brush_color_button = Button(frame1, text="Select Color", command=select_color)
brush_color_button.pack(side=tk.LEFT, padx=10)

frame2 = Frame(root)
frame2.pack(side=tk.TOP, fill=tk.BOTH, expand=True)

canvas = Canvas(frame2, bg='gray')
canvas.pack(fill=tk.BOTH, expand=True)


canvas.bind("<ButtonPress-1>", start_drawing)  
canvas.bind("<B1-Motion>", draw_brush)  
canvas.bind("<ButtonRelease-1>", stop_drawing)  
canvas.bind("<ButtonPress-1>", start_crop)  
canvas.bind("<B1-Motion>", update_crop_area)  
canvas.bind("<ButtonRelease-1>", end_crop)  

root.mainloop()
```
