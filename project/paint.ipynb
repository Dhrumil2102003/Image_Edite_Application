# basic screen creat using tkinter


```python
from tkinter import *
from tkinter.colorchooser import askcolor

root = Tk()
root.title("Paint")

root.geometry("1000x600")

stroke_size = IntVar()
stroke_size.set(1) 

#create stroke_color
stroke_color = StringVar()
stroke_color.set("black")

# Add frame the tkinter
frame1 = Frame(root,height=100, width=1100)
frame1.grid(row=0,column=0, sticky=NW)

# create tool frame
toolFrame = Frame(frame1,height=100,width=100,relief = SUNKEN, borderwidth=3)
toolFrame.grid(row=0,column=0)


# create pencil function
def usepencil():
    stroke_color.set("black")
    canvas['cursor'] = 'pencil'

# create pencil
pencilButton = Button(toolFrame,text="Pencil", width=10,command=usepencil)
pencilButton.grid(row=0,column=0)
    
# create eraser function
def useEraser():
    stroke_color.set("white")
    canvas["cursor"] = "DOTBOX"
    
# create eraser
eraserButton = Button(toolFrame,text="Eraser", width=10, command=useEraser)
eraserButton.grid(row=1,column=0)

# create Tool Label
toolsLabel = Label(toolFrame,text="Tools", width=10, bg = "white")
toolsLabel.grid(row=2,column=0)


# create Size Frame
sizeFrame = Frame(frame1,height=100,width=100,relief = SUNKEN, borderwidth=3)
sizeFrame.grid(row=0,column=1)


# create default button
defaulteButton = Button(sizeFrame,text="Default", width=10, command=useEraser)
defaulteButton.grid(row=0,column=0)

options = [1,2,3,4,5,10]

# create drop down manu
sizeList = OptionMenu(sizeFrame,stroke_size,*options)
sizeList.grid(row=1,column=0)

# create size Label
sizeLabel = Label(sizeFrame,text="Size", width=10, bg = "white")
sizeLabel.grid(row=2,column=0)

# create color box fream
colorBoxFrame = Frame(frame1,height=100,width=100,relief = SUNKEN, borderwidth=3)
colorBoxFrame.grid(row=0,column=2)

def SelectColor():
    selectedColor = askcolor(title="Select Color")
    print(selectedColor)


#create select color button
colorBoxButton = Button(colorBoxFrame,text = "Select Color",width=10,command=SelectColor)
colorBoxButton.grid(row=0,column=0)

frame2 = Frame(root,height=600, width=1100, bg="yellow")
frame2.grid(row=1,column=0)

# Add canvas 
canvas = Canvas(frame2, height=500, width=1100,bg = "white")
canvas.grid(row=0,column=0)


# Variable for pencil 
prvPoint = [0,0]
currentPoint = [0,0]


# create teh function and bind the function
def paint(event):
    global prvPoint, currentPoint
    x =event.x
    y =event.y
    currentPoint = [x,y]
    #canvas.create_line(x,y,120,120,fill="black")# create a line 
#    canvas.create_oval(x,y,x+5,y+5,fill="black")# create a oval 
    
    if prvPoint != [0,0]:
        canvas.create_polygon(prvPoint[0],prvPoint[1],currentPoint[0],currentPoint[1],fill = stroke_color.get(),outline=stroke_color.get(),width= stroke_size.get())
    
    prvPoint = currentPoint
    
    if event.type == "5":
        prvPoint = [0,0]


#canvas.bind("<Button-1>",paint)# many time clicking event
canvas.bind("<B1-Motion>",paint)# one time clicking event
canvas.bind("<ButtonRelease-1>",paint)# one time clicking event



root.resizable(False, False)
root.mainloop()
```
